# Job Prediction Using Logistic Regression

```python
# Import Required  Libraries
import pandas as pd
```

```python
data = pd.read_csv(r'score.csv')
```

```python
data.head()
```

```python
# Label - Admitted
# Features - Others

# X = data.iloc[:,0:3]
# Y = data.iloc[:,3] # No Header is Assigned
```

```python
X = data[['gmat','gpa','work_experience']]
Y = data[['admitted']] # For Better Understanding
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.01)
```

```python
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# Train Model
model.fit(xtrain, ytrain)
```

```python
# Prediction 
test = pd.read_csv(r'stest.csv')
test.head()
```

```python
xpredict = model.predict(test)
```

```python
xpredict
```

```python
test['admitted'] = xpredict
```
